```python
import pandas as pd
```

```python
food_data = pd.read_csv('./CleanFoodData.csv', sep=',')
food_data.rename(columns={'Mean': 'Mean_Food.g'}, inplace=True)
food_data['Mean_CO2.g'] = [0] * len(food_data) # Initialize column for CO2
food_data.head(3)
```

```python
co2_data = pd.read_csv('./CO2Footprint.csv', sep=',', index_col = 0)
co2_data['Grams.CO2e.per.Grams'] = co2_data['Grams.CO2e.per.Serving']/co2_data['Serving.Size'] # Created column for CO2/grams of food
co2_data.head(3)
```

```python
for iFood, foodRow in food_data.iterrows():
    for co2Id, co2Row in co2_data.iterrows():
        if co2Id == foodRow['ID']:
            food_data.loc[iFood,'Mean_CO2.g'] = foodRow['Mean_Food.g']*co2Row['Grams.CO2e.per.Grams']
            break
food_data.head(5)
```

```python
countries = list(food_data['Country'].unique())
ageGroup =list(food_data['Pop_Class'].unique())
countries_df = pd.DataFrame([[0]*3]* len(food_data['Pop_Class'].unique()) * len(food_data['Country'].unique()), columns=["Country", "ageGroup","Mean_CO2.g"])
i = 0
for country in countries:
    for age in ageGroup:
        countries_df.loc[i,'Country'] = country
        countries_df.loc[i,'ageGroup'] = age
        i+=1
        
countries_df
```

```python
for index, countryRow in countries_df.iterrows():
    countries_df.loc[index, 'Mean_CO2.g'] = 0
    for iFood, foodRow in food_data.iterrows():
        if foodRow['Country'] == countryRow['Country'] and foodRow['Pop_Class'] == countryRow['ageGroup']:
            countries_df.loc[index, 'Mean_CO2.g'] = countries_df.loc[index, 'Mean_CO2.g'] + foodRow['Mean_CO2.g']
countries_df
```

```python
countries_df.to_csv("./CO2_per_country_ageGroup.csv", header=True, sep=',', index=False)
```
